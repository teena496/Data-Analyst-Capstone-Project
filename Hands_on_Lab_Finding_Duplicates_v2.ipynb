### Task 2: Analyze Characteristics of Duplicates
1. Identify duplicate rows based on selected columns such as MainBranch, Employment, and RemoteWork. Analyse which columns frequently contain identical values within these duplicate rows.
2. Analyse the characteristics of rows that are duplicates based on a subset of columns, such as MainBranch, Employment, and RemoteWork. Determine which columns frequently have identical values across these rows.
   


```python
## Write your code here
# Find duplicates based on selected columns
dup_subset = df[df.duplicated(subset=['MainBranch', 'Employment', 'RemoteWork'], keep=False)]

# Display how many such duplicates exist
print(f"Number of duplicate rows based on selected columns: {len(dup_subset)}")

# Show a few of them
dup_subset[['MainBranch', 'Employment', 'RemoteWork']].head()
```

```python
# Group by the selected subset columns
grouped = dup_subset.groupby(['MainBranch', 'Employment', 'RemoteWork'])

# Store how often each column has identical values across duplicate groups
from collections import Counter
match_counts = Counter()

# Count the number of groups
total_groups = len(grouped)

# Check uniqueness within each group
for name, group in grouped:
    for col in df.columns:
        if col in ['MainBranch', 'Employment', 'RemoteWork']:
            continue
        if group[col].nunique(dropna=False) <= 1:  # identical (even if NaN)
            match_counts[col] += 1

# Convert counts to percentages
match_freq = {col: (count / total_groups) * 100 for col, count in match_counts.items()}
```

```python
# Sort by frequency
sorted_matches = sorted(match_freq.items(), key=lambda x: x[1], reverse=True)

print("Columns frequently identical in duplicate groups (with % agreement):")
for col, percent in sorted_matches[:10]:  # top 10
    print(f"{col}: {percent:.1f}% of duplicate groups")
```

### Task 3: Visualize Duplicates Distribution
1. Create visualizations to show the distribution of duplicates across different categories.
2. Use bar charts or pie charts to represent the distribution of duplicates by Country and Employment.


```python
## Write your code here
import seaborn as sns

# Find duplicate rows based on subset
dup_df = df[df.duplicated(subset=['MainBranch', 'Employment', 'RemoteWork'], keep=False)]


plt.figure(figsize=(12, 6))
country_counts = dup_df['Country'].value_counts().head(10)  # Top 10 countries
sns.barplot(x=country_counts.index, y=country_counts.values, palette='viridis')
plt.title('Top 10 Countries with Most Duplicate Rows')
plt.xlabel('Country')
plt.ylabel('Number of Duplicate Entries')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
employment_counts = dup_df['Employment'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(employment_counts, labels=employment_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribution of Duplicate Rows by Employment Status')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
def simplify_employment(val):
    if ';' in val:
        return 'Multiple Roles'
    return val

dup_df['SimplifiedEmployment'] = dup_df['Employment'].apply(simplify_employment)
```

```python
employment_counts = dup_df['SimplifiedEmployment'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(employment_counts, labels=employment_counts.index, autopct='%1.1f%%',
        startangle=140, colors=sns.color_palette('pastel'))
plt.title('Simplified Distribution of Duplicate Rows by Employment')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

### Task 4: Strategic Removal of Duplicates
1. Decide which columns are critical for defining uniqueness in the dataset.
2. Remove duplicates based on a subset of columns if complete row duplication is not a good criterion.


```python
## Write your code here
# Define critical columns for uniqueness
subset_cols = ['MainBranch', 'Employment', 'RemoteWork', 'Country', 'YearsCode', 'Age', 'EdLevel']

# Drop duplicates based on these columns
df_cleaned = df.drop_duplicates(subset=subset_cols)

# Check how many were dropped
print(f"Original rows: {df.shape[0]}")
print(f"After removing strategic duplicates: {df_cleaned.shape[0]}")
print(f"Number of rows removed: {df.shape[0] - df_cleaned.shape[0]}")
```
