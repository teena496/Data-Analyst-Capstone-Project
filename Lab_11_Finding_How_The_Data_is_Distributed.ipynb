### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
import matplotlib.pyplot as plt

# Count of each category
job_sat_counts = df['JobSat'].value_counts()

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(job_sat_counts, labels=job_sat_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Job Satisfaction Distribution')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
# Drop NaNs and split language lists
lang_used = df['LanguageHaveWorkedWith'].dropna().str.split(';').explode().str.strip()
lang_wanted = df['LanguageWantToWorkWith'].dropna().str.split(';').explode().str.strip()
```

```python
used_counts = lang_used.value_counts()
wanted_counts = lang_wanted.value_counts()
```

```python
lang_compare = pd.DataFrame({
    'Used': used_counts,
    'Wanted': wanted_counts
}).fillna(0).astype(int)

# Optional: Limit to top 10 by usage
lang_compare_top = lang_compare.sort_values('Used', ascending=False).head(10)

# Plot it
import matplotlib.pyplot as plt

lang_compare_top.plot(kind='bar', figsize=(12, 6))
plt.title('Programming Languages: Used vs. Wanted')
plt.xlabel('Programming Language')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
remote_region_df = df[['Country', 'RemoteWork']].dropna()
```

```python
remote_counts = pd.crosstab(remote_region_df['Country'], remote_region_df['RemoteWork'])
```

```python
remote_percent = pd.crosstab(remote_region_df['Country'], remote_region_df['RemoteWork'], normalize='index') * 100
```

```python
# Filter top 10 countries by total respondents
top_countries = remote_region_df['Country'].value_counts().nlargest(10).index
remote_top = remote_counts.loc[top_countries]

# Plot
remote_top.plot(kind='bar', figsize=(12, 6))
plt.title('Remote Work Preference by Country (Top 10)')
plt.xlabel('Country')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.legend(title='Remote Work')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Choose top 10-15 countries
top_countries = remote_percent.sum(axis=1).nlargest(15).index
remote_heatmap = remote_percent.loc[top_countries]

plt.figure(figsize=(12, 8))
sns.heatmap(remote_heatmap, annot=True, fmt=".1f", cmap="coolwarm", cbar_kws={'label': '%'})
plt.title('Remote Work Distribution by Country (Normalized %)')
plt.xlabel('Remote Work Type')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here
import numpy as np
df['YearsCodePro_cleaned'] = df['YearsCodePro'].replace({
    'Less than 1 year': 0.5,
    'More than 50 years': 51
})
df['YearsCodePro_cleaned'] = pd.to_numeric(df['YearsCodePro_cleaned'], errors='coerce')
df['YearsCodePro_cleaned'].fillna(np.nan, inplace=True)
```

```python
df['YearsCodePro_cleaned'].dropna(inplace=True)
```

```python
df['JobSat'].fillna(np.nan)
```

```python
df['JobSat'].dropna()
```

```python
df['JobSat'].value_counts()
```

```python
df_corr = df.dropna(subset=['YearsCodePro_cleaned', 'JobSat'])
```

```python
# Pearson (linear correlation)
!pip install scipy

pearson_corr = df_corr['YearsCodePro_cleaned'].corr(df_corr['JobSat'], method='pearson')

# Spearman (rank-based correlation)
spearman_corr = df_corr['YearsCodePro_cleaned'].corr(df_corr['JobSat'], method='spearman')

print(f"Pearson Correlation: {pearson_corr:.3f}")
print(f"Spearman Correlation: {spearman_corr:.3f}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.regplot(x='YearsCodePro_cleaned', y='JobSat', data=df_corr, scatter_kws={'alpha':0.4})
plt.title('Job Satisfaction vs Years of Experience')
plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Job Satisfaction (1 = Very Dissatisfied, 5 = Very Satisfied)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
df['EdLevel'].replace({
       'Primary/elementary school':'Primary school',
       'Bachelor’s degree (B.A., B.S., B.Eng., etc.)':'Bachelor’s degree',
       'Master’s degree (M.A., M.S., M.Eng., MBA, etc.)':'Master’s degree',
       'Some college/university study without earning a degree' : 'University study without degree',
       'Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)':'Secondary school',
       'Professional degree (JD, MD, Ph.D, Ed.D, etc.)':'Professional degree',
       'Associate degree (A.A., A.S., etc.)':'Associate degree'},inplace=True)
df['EdLevel'].dropna(inplace=True)
```

```python
df['EdLevel'].unique()
df['Employment'].unique()



edlevel_emp_df = df[['EdLevel','Employment']].dropna()
edlevel_emp_df
```

```python
edlevel_emp_df = edlevel_emp_df.assign(
    Employment_Types=edlevel_emp_df['Employment'].str.split(';')
).explode('Employment_Types')
edlevel_emp_df
```

```python
## Write your code here
import pandas as pd

# Generate the cross-tab table
employment_edlevel_ct = pd.crosstab(edlevel_emp_df['EdLevel'], edlevel_emp_df['Employment_Types'])
```

```python
import matplotlib.pyplot as plt

# Plot the cross-tab as a stacked bar chart
employment_edlevel_ct.plot(kind='bar', stacked=True, figsize=(12, 6))

plt.title('Employment Status by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.legend(title='Employment Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
employment_edlevel_percent = pd.crosstab(edlevel_emp_df['EdLevel'], edlevel_emp_df['Employment_Types'], normalize='index') * 100

employment_edlevel_percent.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='viridis')

plt.title('Employment Status by Education Level (Percentage)')
plt.xlabel('Education Level')
plt.ylabel('Percentage of Respondents')
plt.xticks(rotation=45)
plt.legend(title='Employment Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
