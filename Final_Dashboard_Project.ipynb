```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('survey_data_updated 5.csv')
df.head()
```

```python
df['LanguageHaveWorkedWith'].unique()
```

```python
df = df.assign(
    Languages = df['LanguageHaveWorkedWith'].str.split(';')
).explode('Languages')

```

```python
df['Languages'].unique()
```

```python
df.to_csv("language_expanded.csv", index=False)
```

```python
# Expand LanguageHaveWorkedWith column
expanded_rows = []
for idx, row in df.iterrows():
    respondent_id = row.get("ResponseId", idx)  # fallback to index if no ID column
    languages = row["LanguageHaveWorkedWith"]
    if pd.notna(languages):
        for lang in languages.split(";"):
            expanded_rows.append({"ResponseId": respondent_id, "Language": lang.strip()})

# Create new DataFrame and save to CSV
language_expanded_df = pd.DataFrame(expanded_rows)
language_expanded_df.to_csv("language_expanded.csv", index=False)
```

```python
import pandas as pd

df = pd.read_csv("survey_data_updated 5.csv")

db_rows = []
for idx, row in df.iterrows():
    respondent_id = row.get("ResponseId", idx)
    databases = row["DatabaseHaveWorkedWith"]
    if pd.notna(databases):
        for db in databases.split(";"):
            db_rows.append({"ResponseId": respondent_id, "Database": db.strip()})

db_df = pd.DataFrame(db_rows)
db_df.to_csv("database_expanded.csv", index=False)
```

```python
platform_rows = []
for idx, row in df.iterrows():
    respondent_id = row.get("ResponseId", idx)
    platforms = row["PlatformHaveWorkedWith"]
    if pd.notna(platforms):
        for p in platforms.split(";"):
            platform_rows.append({"ResponseId": respondent_id, "Platform": p.strip()})

platform_df = pd.DataFrame(platform_rows)
platform_df.to_csv("platform_expanded.csv", index=False)
```

```python
webframe_rows = []
for idx, row in df.iterrows():
    respondent_id = row.get("ResponseId", idx)
    frameworks = row["WebframeHaveWorkedWith"]
    if pd.notna(frameworks):
        for f in frameworks.split(";"):
            webframe_rows.append({"ResponseId": respondent_id, "WebFramework": f.strip()})

webframe_df = pd.DataFrame(webframe_rows)
webframe_df.to_csv("webframe_expanded.csv", index=False)
```

```python
import pandas as pd

df = pd.read_csv("survey_data_updated 5.csv")

# Initialize empty lists
lang_want_rows = []
db_want_rows = []
platform_want_rows = []
webframe_want_rows = []

# Process each row
for idx, row in df.iterrows():
    respondent_id = row.get("ResponseId", idx)

    # Expand LanguageWantToWorkWith
    if pd.notna(row.get("LanguageWantToWorkWith")):
        for lang in row["LanguageWantToWorkWith"].split(";"):
            lang_want_rows.append({"ResponseId": respondent_id, "LanguageWantToWorkWith": lang.strip()})

    # Expand DatabaseWantToWorkWith
    if pd.notna(row.get("DatabaseWantToWorkWith")):
        for db in row["DatabaseWantToWorkWith"].split(";"):
            db_want_rows.append({"ResponseId": respondent_id, "DatabaseWantToWorkWith": db.strip()})

    # Expand PlatformWantToWorkWith
    if pd.notna(row.get("PlatformWantToWorkWith")):
        for platform in row["PlatformWantToWorkWith"].split(";"):
            platform_want_rows.append({"ResponseId": respondent_id, "PlatformWantToWorkWith": platform.strip()})

    # Expand WebframeWantToWorkWith
    if pd.notna(row.get("WebframeWantToWorkWith")):
        for webframe in row["WebframeWantToWorkWith"].split(";"):
            webframe_want_rows.append({"ResponseId": respondent_id, "WebframeWantToWorkWith": webframe.strip()})

# Save to CSV
pd.DataFrame(lang_want_rows).to_csv("language_want_expanded.csv", index=False)
pd.DataFrame(db_want_rows).to_csv("database_want_expanded.csv", index=False)
pd.DataFrame(platform_want_rows).to_csv("platform_want_expanded.csv", index=False)
pd.DataFrame(webframe_want_rows).to_csv("webframe_want_expanded.csv", index=False)
```
