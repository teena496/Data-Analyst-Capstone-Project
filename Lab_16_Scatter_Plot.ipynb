#### Step 1: Load the dataset


```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

df = pd.read_csv(file_path)
```


### Task 1: Exploring Relationships with Scatter Plots



#### 1. Scatter Plot for Age vs. Job Satisfaction



Visualize the relationship between respondents' age (`Age`) and job satisfaction (`JobSatPoints_6`). Use this plot to identify any patterns or trends.




```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert Age to numeric values if not already done
age_map = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 45,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}

# Apply the mapping if Age column is categorical
df['Age_numeric'] = df['Age'].map(age_map)

# Filter out rows with missing values in Age or JobSatPoints_6
df_filtered = df[['Age_numeric', 'JobSatPoints_6']].dropna()

# Plotting the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_filtered, x='Age_numeric', y='JobSatPoints_6', color='blue')

# Title and labels
plt.title('Scatter Plot of Age vs Job Satisfaction', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=12)
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Compensation vs. Job Satisfaction


Explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) using a scatter plot.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Filter out rows with missing values in ConvertedCompYearly or JobSatPoints_6
df_filtered = df[['ConvertedCompYearly', 'JobSatPoints_6']].dropna()

# Plotting the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_filtered, x='ConvertedCompYearly', y='JobSatPoints_6', color='green')

# Title and labels
plt.title('Scatter Plot of Compensation vs Job Satisfaction', fontsize=14)
plt.xlabel('Yearly Compensation (ConvertedCompYearly)', fontsize=12)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=12)
plt.tight_layout()
plt.show()
```

### Task 2: Enhancing Scatter Plots


#### 1. Scatter Plot with Trend Line for Age vs. Job Satisfaction



Add a regression line to the scatter plot of Age vs. JobSatPoints_6 to highlight trends in the data.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert Age to numeric values if not already done
age_map = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 45,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}

# Apply the mapping if Age column is categorical
df['Age_numeric'] = df['Age'].map(age_map)

# Filter out rows with missing values in Age or JobSatPoints_6
df_filtered = df[['Age_numeric', 'JobSatPoints_6']].dropna()

# Plotting the scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(data=df_filtered, x='Age_numeric', y='JobSatPoints_6', scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})

# Title and labels
plt.title('Scatter Plot with Trend Line: Age vs Job Satisfaction', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=12)
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Age vs. Work Experience


Visualize the relationship between Age (`Age`) and Work Experience (`YearsCodePro`) using a scatter plot.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming the 'Age' column has been converted to numeric as done previously
# If not, use the same mapping as earlier
age_map = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 45,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}


# Apply the mapping if Age column is categorical
df['Age_numeric'] = df['Age'].map(age_map)

# Convert YearsCodePro to numeric (some values might be strings like 'Less than 1 year')
def convert_years(x):
    if x == 'Less than 1 year':
        return 0.5
    elif x == 'More than 50 years':
        return 51
    try:
        return float(x)
    except:
        return None

df['YearsCodePro_numeric'] = df['YearsCodePro'].apply(convert_years)

# Filter out rows with missing values in Age or YearsCodePro
df_filtered = df[['Age_numeric', 'YearsCodePro_numeric']].dropna()

# Plotting the scatter plot
plt.figure(figsize=(6, 6))
sns.scatterplot(data=df_filtered, x='Age_numeric', y='YearsCodePro_numeric', color='purple')

# Title and labels
plt.title('Scatter Plot of Age vs Work Experience', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Years of Work Experience (YearsCodePro)', fontsize=12)
plt.tight_layout()
plt.show()
```

### Task 3: Combining Scatter Plots with Additional Features


#### 1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size



Create a bubble plot to explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), with bubble size representing age.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming the 'Age' column has already been converted to numeric values as done previously
# Apply mapping for Age (if not already done)
age_map = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 45,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70
}

df['Age_numeric'] = df['Age'].map(age_map)

# Filter out rows with missing values for the relevant columns
df_filtered = df[['ConvertedCompYearly', 'JobSatPoints_6', 'Age_numeric']].dropna()

# Plotting the bubble plot
plt.figure(figsize=(12, 8))
plt.scatter(data=df_filtered,
            x='ConvertedCompYearly',  # X-axis: Yearly Compensation
            y='JobSatPoints_6',       # Y-axis: Job Satisfaction
            s=df_filtered['Age_numeric']*10,  # Bubble size: Age (scaled for visibility)
            alpha=0.5,                # Transparency of bubbles
            color='skyblue',          # Color of the bubbles
            edgecolors="w",           # Edge color of bubbles
            linewidth=0.5)            # Edge line width

# Title and labels
plt.title('Bubble Plot of Compensation vs Job Satisfaction with Age as Bubble Size', fontsize=14)
plt.xlabel('Yearly Compensation (ConvertedCompYearly)', fontsize=12)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=12)
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Popular Programming Languages by Job Satisfaction


Visualize the popularity of programming languages (`LanguageHaveWorkedWith`) against job satisfaction using a scatter plot. Use points to represent satisfaction levels for each language.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the dataset has been loaded into 'df'
# Example: Let's assume LanguageHaveWorkedWith is a list of languages

# Preprocess the data (expand languages and calculate average job satisfaction)
df_languages = df[['LanguageHaveWorkedWith', 'JobSatPoints_6']].dropna()

# We need to "explode" the list of languages into individual rows
languages_exploded = df_languages.assign(
    Languages=df['LanguageHaveWorkedWith'].str.split(';')
).explode('Languages')

# Group by language and calculate the average job satisfaction
avg_satisfaction_per_language = languages_exploded.groupby('Languages')['JobSatPoints_6'].mean().reset_index()

# Plotting the scatter plot
plt.figure(figsize=(12, 8))
plt.scatter(data=avg_satisfaction_per_language,
            x='Languages',  # X-axis: Programming Language
            y='JobSatPoints_6',          # Y-axis: Average Job Satisfaction
            color='skyblue',             # Color of the points
            edgecolors="w",              # Edge color of points
            linewidth=0.5,               # Edge line width
            alpha=0.7)                   # Transparency of points

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha="right")

# Title and labels
plt.title('Scatter Plot of Programming Languages vs Job Satisfaction', fontsize=14)
plt.xlabel('Programming Languages', fontsize=12)
plt.ylabel('Average Job Satisfaction (JobSatPoints_6)', fontsize=12)
plt.tight_layout()
plt.show()
```

### Task 4: Scatter Plot Comparisons Across Groups


#### 1. Scatter Plot for Compensation vs. Job Satisfaction by Employment Type


Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), categorized by employment type (`Employment`). Use color coding or markers to differentiate between employment types.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming the dataset is loaded into 'df'

# Check for missing values and drop if necessary
df_clean = df.dropna(subset=['ConvertedCompYearly', 'JobSatPoints_6', 'Employment'])

df_clean = df_clean.assign(
    Employment_Types=df['Employment'].str.split(';')
).explode('Employment_Types')

# Plotting the scatter plot with color coding by Employment Type
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df_clean,
                x='ConvertedCompYearly',           # X-axis: Yearly Compensation
                y='JobSatPoints_6',                # Y-axis: Job Satisfaction
                hue='Employment_Types',                  # Color by Employment Type
                style='Employment_Types',                # Different markers for each Employment Type
                palette='Set1',                    # Color palette for different employment types
                s=100,                             # Size of points
                edgecolor='black',                 # Edges for the points
                alpha=0.7)                         # Transparency of points

# Title and labels
plt.title('Compensation vs Job Satisfaction by Employment Type', fontsize=14)
plt.xlabel('Yearly Compensation (ConvertedCompYearly)', fontsize=12)
plt.ylabel('Job Satisfaction (JobSatPoints_6)', fontsize=12)

# Show legend
plt.legend(title='Employment Type', loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Work Experience vs. Age Group by Country


Compare work experience (`YearsCodePro`) across different age groups (`Age`) and countries (`Country`). Use colors to represent different countries and markers for age groups.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Filter out rows with 'Under 18 years old' and 'Prefer not to say' in the Age column
df_clean = df.dropna(subset=['YearsCodePro', 'Age', 'Country'])

# Remove unwanted categories from 'Age'
df_clean = df_clean[~df_clean['Age'].isin(['Under 18 years old', 'Prefer not to say'])]
# Convert YearsCodePro to numeric (some values might be strings like 'Less than 1 year')
def convert_years(x):
    if x == 'Less than 1 year':
        return 0.5
    elif x == 'More than 50 years':
        return 51
    try:
        return float(x)
    except:
        return None

df_clean['YearsCodePro_numeric'] = df_clean['YearsCodePro'].apply(convert_years)

# Plotting the scatter plot with color coding by Country and markers for Age
plt.figure(figsize=(12, 10))

sns.scatterplot(data=df_clean,
                x='YearsCodePro_numeric',                # X-axis: Years of Work Experience
                y='Age',                         # Y-axis: Age group
                hue='Country',                   # Color by Country
                style='Age',                     # Different markers for each Age group
                palette='Set1',                   # Color palette for countries
                markers={'18-24 years old': 'o',
                         '25-34 years old': 's',
                         '35-44 years old': '^',
                         '45-54 years old': 'D',
                         '55-64 years old': 'P',
                         '65 years or older': 'X'}, # Markers for different age groups
                s=100,                            # Size of points
                edgecolor='black',                # Edges for the points
                alpha=0.7)                        # Transparency of points

# Title and labels
plt.title('Work Experience vs Age Group by Country', fontsize=14)
plt.xlabel('Years of Work Experience (YearsCodePro)', fontsize=12)
plt.ylabel('Age Group', fontsize=12)

# Show legend
plt.legend(title='Country', loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.show()
```
